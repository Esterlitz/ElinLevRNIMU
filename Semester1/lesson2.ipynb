```python
a=int(input())
b=int(input())
for j in range(a, b+1):
    s=0
    for i in range(2, j):
        if j%i==0:
        s+=1
    if s==0:
        print(s)
```

```python
x=int(input('Вложенная сумма: '))
p=int(input('Процент: '))
y=int(input('Желаемая сумма: '))
years=0
while x<y:
    x=x*(1+p/100)
    x=int(x//1)
    years+=1
print(years)
```

```python
m=0
sec=0
h=int(input())
while h!=0:
    if h>m:
        sec=m
        m=h
    h=int(input())
print(sec)
```
