```python
alphabet=set('abcdefghijklmnopqrstuvwxyz')
A=set(input("Первая строка: "))
B=set(input("Вторая строка: "))
print("Есть в первой, но не во второй: " + ", ".join(A-B))
print("Есть во второй, но не в первой: " + ", ".join(B-A))
print('Нет в первой: ' + ", ".join(alphabet-A))
print('Нет во второй: ' + ", ".join(alphabet-B))
```

```python
vetbase=[("Дуся", 4, "Иван", "Иванович"), ("Мурзик", 7, "Аркадий", "Паровозов"), ("Барсик", 3, "Иван", "Иванович")]
owners={}
for cat, age, name, surname in vetbase:
    full_name=name+' '+surname
    if full_name not in owners:
        owners[full_name]=[]
    owners[full_name].append(cat)
    owners[full_name].append(age)
userword=input()
print(userword+": "+', '.join(map(str, owners[userword])))
```

```python
from time import time
import random
H=[random.randint(1, 100) for i in range(int(input("Сколько хотите чисел?")))]
print(H)
l=len(H)
time_start=time()
for j in range(l-1):
    for i in range(l-1-j):
        if H[i] > H[i+1]:
            H[i], H[i+1] = H[i+1], H[i]
print(H)
print(time()-time_start)
```

```python
import random
kol1=[]
kol2=[]
kol3=[]
for i in range(1, 8):
    mix=random.randint(1, 3)
    if mix==1:
        kol1.append(i)
    elif mix==2:
        kol2.append(i)
    elif mix==3:
        kol3.append(i)
random.shuffle(kol1)
random.shuffle(kol2)
random.shuffle(kol3)
print("Правила игры:")
print('Перемещай цифры указывая номер колышка, с помощью клавишь 1, 2 или 3.')
print("Если хочешь сдаться, пиши 0.")
print("Количество попыток не ограничено")
while True:
    print()
    print('Кольца на первом колышке: ' + ', '.join(map(str, kol1)))
    print('Кольца на втором колышке: ' + ', '.join(map(str, kol2)))
    print('Кольца на третьем колышке: ' + ', '.join(map(str, kol3)))
    print()
    if kol1==[7, 6, 5, 4, 3, 2, 1] or kol2==[7, 6, 5, 4, 3, 2, 1] or kol3==[7, 6, 5, 4, 3, 2, 1]:
        print("Ты победил!")
        break
    kolfrom=input("С какого колышка снимешь кольцо? ")
    if kolfrom=='0':
        print("Ты сдался!")
        break
    kolto=input("На какой колышек наденешь кольцо? ")
    if kolfrom=='1' and kolto=='2':
        if len(kol1)==0:
            print("Так нельзя!")
        else:
            top=kol1.pop(-1)
            kol2.append(top)
    elif kolfrom=='1' and kolto=='3':
        if len(kol1)==0:
            print("Так нельзя!")
        else:
            top=kol1.pop(-1)
            kol3.append(top)
    elif kolfrom=='2' and kolto=='1':
        if len(kol2)==0:
            print("Так нельзя!")
        else:
            top=kol2.pop(-1)
            kol1.append(top)
    elif kolfrom=='2' and kolto=='3':
        if len(kol2)==0:
            print("Так нельзя!")
        else:
            top=kol2.pop(-1)
            kol3.append(top)
    elif kolfrom=='3' and kolto=='1':
        if len(kol3)==0:
            print("Так нельзя!")
        else:
            top=kol3.pop(-1)
            kol1.append(top)
    elif kolfrom=='3' and kolto=='2':
        if len(kol3)==0:
            print("Так нельзя!")
        else:
            top=kol3.pop(-1)
            kol2.append(top)
    elif kolto==0:
        print("Ты сдался!")
    else:
        print("Так нельзя!")
```
